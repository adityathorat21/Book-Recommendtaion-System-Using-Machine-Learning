## Popularity based recommender system

```python
rating_merge = ratings.merge(books,on='ISBN')
rating_merge.head()
```

```python
grp_rating= rating_merge.groupby('Book-Title').count()['Book-Rating'].reset_index()
grp_rating.rename(columns={'Book-Rating':'rating-Count'}, inplace=True)
grp_rating
```

```python
avg_rating= rating_merge.groupby('Book-Title').mean(numeric_only = True)['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg-rating'}, inplace=True)
avg_rating.head()
```

```python
popular= grp_rating.merge(avg_rating,on='Book-Title')
popular
```

```python
popular50=popular[popular['rating-Count']>=250].sort_values('avg-rating', ascending=False).head(50)
popular50
```

```python
popular_books=popular50.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','rating-Count','avg-rating','Image-URL-M']]
popular_books
```

## Collaborative Filtering Based recommender system

```python
x=rating_merge.groupby('User-ID').count()['Book-Rating']>200
rated200_users=x[x].index
rated200_users
```

```python
filtered_users = rating_merge[rating_merge['User-ID'].isin(rated200_users)]
filtered_users
```

```python
y=filtered_users.groupby('Book-Title').count()['Book-Rating'] >= 50
rate50_books=y[y].index
rate50_books
```

```python
filtered_rating = filtered_users[filtered_users['Book-Title'].isin(rate50_books)]
filtered_rating
```

```python
data_table = filtered_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
data_table
```

```python
data_table.fillna(0, inplace=True)
data_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(data_table)
similarity_score
```

```python
def recommend(book_name):
    index=np.where(data_table.index==book_name)[0][0]
    similar_books=sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]

    data = []
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title'] == data_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('Animal Farm')
```
